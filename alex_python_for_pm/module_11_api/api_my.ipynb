```python
domain = 'google.com'
url = f'https://api.domainsdb.info/v1/domains/search?domain={domain}'
# смотрим купленные домены гугла

res = requests.get(url)
res.json() #смотрим боди ответа
```

```python
domains = res.json() #специальный метод, который возврвщает боди ответа (текст) в словарь из джейсона
domains
```

```python
[d['domain'] for d in domains['domains']] #найди все значения для ключа domain
```

```python
res = requests.get('https://api.ipify.org?format=json') #возвращает твой айпи
ip = res.json()['ip'] #достаем значение по ключу айпи
ip
```

```python
res = requests.get(f'https://ipinfo.io/{ip}/geo') #можем передать айпи и он передаст данные о геолокации
```

```python
res.json()
```

```python
# post

url = 'https://csclub.uwaterloo.ca/~phthakka/1pt-express/addURL'

params = {
    'long': 'http://google.com',
    'short': 'sunnybunny'
}

res = requests.post(url, params=params) #у метода requests.post есть параметр params! тут просто совпало, что параметр называется также
res

#дальше перешли на страницу 1pt.com/sunnybunny → переадресовало на google.com
```
