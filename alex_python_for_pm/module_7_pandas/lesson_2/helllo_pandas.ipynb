```python
users.to_csv('users2.csv') #сохранит новый файл с названием users2
```

```python
c = users.to_dict() #сохранит в виде словаря
print(c) #выведет словарь
```

```python
c = users.to_string(index=None) #сохранит в виде текста, индекс уберет
print(c) #выведет текст
```

```python
c = users.to_csv(index=None) #сохранит в виде csv
print(c) #выведет csv
```
