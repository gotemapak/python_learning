```python
import pandas as pd
```

```python
df = pd.read_csv('ads.csv')
df.head()
```

```python
# расходы на каждый рекл канал в каждой стране
# стоимость клика в каждой стране и каждом рекламном канале
# данные были читаемыми
```

```python
df.pivot_table(
    index='country', #строки → можем заполнить списком
    columns= 'ad_channel', #колонки → можем заполнить списком
    values='ad_spend', #значения → можем заполнить списком
    aggfunc=['sum','count'] #функция → можем заполнить списком
) #удалить NaN можно через .dropna() или fillna(0) #заменить на нули
```

```python
df.set_index('country').loc['US'][['date', 'ad_channel', 'impressions']]
```

```python
(
    df
    .set_index('country') #установили индекс на страну
    .loc['US'] #взяли индексы со значением US
    .pivot( #сделали пивот на основе данных ниже
        index='date', #установили значения для строк
        columns='ad_channel', #выбрали значения для колонок
        values='impressions' #взяли в качестве value impressions
    )
    .fillna(0) #поставили вместо nan 0
    .astype(int) #установили вывод в качестве целочисленных
) #этот способ используется чтобы переоформлить дата фрейм из длинного в широкий
```

```python
av_spend = df.groupby(['country', 'ad_channel'])['ad_spend'].sum().mean() 
#средние расходы по странам
# print(av_spend)

def get_av(values):
    x = sum(values)
    if x < av_spend:
        return 'below_av'
    elif x == av_spend:
        return 'av'
    else:
        return 'above_av'

df.pivot_table(
    index='country', 
    columns= 'ad_channel', 
    values='ad_spend', 
    aggfunc=get_av 
) 

# below av
# av
# above av
```
